# Lists in Python

Lists in Python are one of the 4 built-in data types that we can use to store collections of data. 
For example:

****Creating Ã  list in Python****

```python
list1 = ["Python", "SQL", "R", "Julia", "Scala", "JS"]
```

***Lists are changeable, ordered and can take dublicate values***

We can print out the values of the list by using print

```python
print(list1)
```

## Get a random Element from a list

```python
import random 
random.choice(list1)
```

## Get multiple random elements from a list

Suppose i want to get randomly 3 elements from my list.

```python
import random
random.sample(list1, 3)
```

## We can add a list to our existing list by using the Append method.

```python
list1.append(["Java","JQuery", "C++"])
```

```python
print(list1)
```

## We can add elements to our list by using Extend methode

```python
list1.extend(["Java","JQuery", "C++"])
```

```python
print(list1)
```

## Counting number of elements of the list

We can count the number of elements of a list using **Len** function.

```python
print(len(list1))
```

# Dictionaries in Python

**In Python, dictionaries are used to store data in key:value pairs. Unlike lists, dictionaries in Python do not accept duplicates.**

A dictionary can store any kind of data type.

```python
vehicle = {
  "brand": "Toyota",
  "model": "CHR",
  "hybrid": True,
  "year": 2018,
  "colors": ["red", "white", "blue"]
}
```

```python
print(vehicle)
```

## Example of Dictionary in real word project

In the example below, a have a column (**Primarypropertytype**) in my dataset with many values. I have 32 types of Appartments and i create a dictionary to group these 32 types of appartments into 7 groups.

```python
data  = pd.read_csv("https://raw.githubusercontent.com/cheikhb/P3_Openclass_IML/main/seattle_df.csv")
```

```python
data.head()
```

```python
data.Primarypropertytype.unique()
```

```python
data.Primarypropertytype.nunique()
```

The dictionary 

```python
apt_type = {'Hotel' : 'Housing', 'Other' : 'Other', 'Mid-Rise Multifamily' : 'Housing', 'Mixed Use Property' : 'Housing',
       'K-12 School' : 'Education', 'College/University' : 'Education', 'Small- and Mid-Sized Office' : 'Office',
       'Self-Storage Facility\n' : 'Office', 'Distribution Center' : 'Office', 'Large Office' : 'Office',
       'Retail Store' : 'Retail', 'Low-Rise Multifamily' : 'Housing', 'Senior Care Community' : 'Housing',
       'Medical Office' : 'Health', 'Hospital' : 'Health', 'Residence Hall/Dormitory' : 'Housing',
       'Non-Refrigerated Warehouse' : 'Housing', 'Distribution Center\n' : 'Office',
       'SPS-District K-12' : 'Housing', 'Worship Facility' : 'Leisure',
       'Supermarket/Grocery Store' : 'Retail', 'Laboratory' : 'Office', 'Self-Storage Facility' : 'Office',
       'Refrigerated Warehouse' : 'Housing', 'High-Rise Multifamily' : 'Housing', 'Restaurant\n' : 'Leisure',
       'Restaurant' : 'Leisure', 'University' : 'Education', 'Warehouse' : 'Office', 'Residence Hall' : 'Housing',
       'Supermarket / Grocery Store' : 'Retail', 'Office' : 'Office'}
```

I can now use the Map function to group all these 32 values of the Primarypropertytype column into only 7 values for this column in my dataset.

```python
data['Primarypropertytype'] = data['Primarypropertytype'].map(apt_type)
```

```python
data['Primarypropertytype'].unique()
```

And now we go from 32 differents properties types to only 7.
